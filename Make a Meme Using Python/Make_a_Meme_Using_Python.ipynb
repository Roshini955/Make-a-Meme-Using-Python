# Creating a Programming Meme

```python
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt

# Create a new image with a white background
width, height = 600, 500
image = Image.new('RGB', (width, height), color='white')

# Create a drawing object
draw = ImageDraw.Draw(image)

# Add a simple background
draw.rectangle([0, 0, width, height], fill='lightblue')

# Add some text
text = "When you finally fix that bug"
font_size = 40

# Attempt to use a different font path or a default font
try:
    font = ImageFont.truetype("arial.ttf", font_size)
except IOError:
    font = ImageFont.load_default()

# Calculate text position to center it
text_bbox = draw.multiline_textbbox((0, 0), text, font=font)
text_position = ((width - text_bbox[2]) / 2, 50)

# Draw the text
draw.multiline_text(text_position, text, font=font, fill='black', align='center')

# Add a simple image (let's draw a smiley face)
face_center = (width // 2, height - 200)
face_radius = 150
draw.ellipse([face_center[0] - face_radius, face_center[1] - face_radius,
              face_center[0] + face_radius, face_center[1] + face_radius],
              fill='yellow', outline='black', width=5)

# Add eyes
eye_radius = 20
left_eye = (face_center[0] - 60, face_center[1] - 40)
right_eye = (face_center[0] + 60, face_center[1] - 40)
draw.ellipse([left_eye[0] - eye_radius, left_eye[1] - eye_radius,
              left_eye[0] + eye_radius, left_eye[1] + eye_radius], fill='black')
draw.ellipse([right_eye[0] - eye_radius, right_eye[1] - eye_radius,
              right_eye[0] + eye_radius, right_eye[1] + eye_radius], fill='black')

# Add a smile
smile_bbox = [face_center[0] - 80, face_center[1] - 10,
              face_center[0] + 80, face_center[1] + 80]
draw.arc(smile_bbox, start=0, end=180, fill='black', width=10)

# Save the image
output_filename = 'programming_meme_fixed.png'
image.save(output_filename)
print(f"Meme created and saved as {output_filename}")

# Display the image using matplotlib
plt.figure(figsize=(7, 5))
plt.imshow(image)
plt.axis('off')
plt.show()
```

# Another Meme


```python
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt

# Open the local image
img = Image.open("img/meme_1.png")


# Create a drawing object
draw = ImageDraw.Draw(img)

# Use a default font
font = ImageFont.load_default()

# Add text to the image with larger font size
draw.text((10, 10), "When you realize your dataset has missing values", fill="black", font=font)
draw.text((img.width - 150, img.height - 100), "And it's time for EDA", fill="black", font=font)

# Save the meme
img.save("eda_meme.png")

# Display the meme
plt.figure(figsize=(10, 7))
plt.imshow(img)
plt.axis('off')
plt.show()

print("EDA meme created and saved as 'eda_meme.png'")
```

# Tutorial Meme


```python
from PIL import Image, ImageDraw, ImageFont

# Load the image
img = Image.open('img/meme_2.png')

# Initialize the drawing context
draw = ImageDraw.Draw(img)


text = "When you realize you just watched all the tutorials without actually coding"

# Load the font
try:
    font = ImageFont.truetype("DejaVuSans-Bold.ttf", 36)
except IOError:
    font = ImageFont.load_default()

# Calculate text size and position using textbbox for Pillow >= 8.0.0
bbox = draw.textbbox((0, 0), text, font)
textwidth = bbox[2] - bbox[0]
textheight = bbox[3] - bbox[1]

width, height = img.size
x = (width - textwidth) / 2
y = height - textheight - 20

# Add text to the image
draw.text((x, y), text, font=font, fill="white", stroke_width=2, stroke_fill="black")

# Save the meme
meme_filename = 'tutorial_meme.png'
img.save(meme_filename)

print(f"Meme created and saved as {meme_filename}")
display(img)
```
